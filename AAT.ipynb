## Importing Libraries and Creating Dataframes
---

```python
## Standard Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Importing the OS and JSON Modules
import os,json
```

```python
## First Dataframe
df_superheros = pd.read_csv("Data/superhero_info - superhero_info.csv")
df_superheros.head()
```

```python
## Second Dataframe
df_powers = pd.read_csv("Data/superhero_powers - superhero_powers.csv")
df_powers.head()
```

## Formatting First Dataframe
---

```python
## Seperating Hero|Publisher into multiple columns
df_superheros[["Hero", "Publisher"]] = df_superheros["Hero|Publisher"].str.split("|", expand = True)

## Dropping "Hero|Publisher" columns
df_superheros.drop(columns = "Hero|Publisher", inplace = True)

## Displaying data to verify data has been transformed correctly
df_superheros.head()
```

```python
## Seperating Measurements into multiple columns
df_superheros[["Height", "Weight"]] = df_superheros["Measurements"].str.split(",", expand = True)

## Dropping Measurements columns
df_superheros.drop(columns = "Measurements", inplace = True)

## Displaying data to verify data has been transformed correctly
df_superheros.head()
```

```python
## Seperating Height into multiple columns to isolate numeric value
df_superheros[["Label", "Height", "Unit"]] = df_superheros["Height"].str.split(" ", expand = True)

## Dropping Unneeded columns
df_superheros.drop(columns = "Label", inplace = True)
df_superheros.drop(columns = "Unit", inplace = True)

## Displaying data to verify data has been transformed correctly
df_superheros.head()
```

```python
## Seperating Height into multiple columns to isolate numeric value
df_superheros[["1", "Label", "Weight", "Unit"]] = df_superheros["Weight"].str.split(" ", expand = True)

## Dropping Unneeded columns
df_superheros.drop(columns = "Label", inplace = True)
df_superheros.drop(columns = "Unit", inplace = True)
df_superheros.drop(columns = "1", inplace = True)

## Displaying data to verify data has been transformed correctly
df_superheros.head()
```

```python
## Removing ' from Height and Weight and converting to float

## Value to replace
to_replace = "'"

## Creating loop to replace all '
for char in to_replace:
    df_superheros["Height"] = df_superheros["Height"].str.replace(char, "", regex = False)
    df_superheros["Weight"] = df_superheros["Weight"].str.replace(char, "", regex = False)

## Saving data type as float for both columns
df_superheros["Height"] = df_superheros["Height"].astype(float)
df_superheros["Weight"] = df_superheros["Weight"].astype(float)

## Displaying data to verify data has been transformed correctly
df_superheros.info()
```

## Formatting Second Dataframe
---

```python
## One Hot encoding the powers column
df_powers["Powers_split"] = df_powers["Powers"].str.split(",")
df_powers["Powers_split"].head()

exploded = df_powers.explode("Powers_split")
columns = exploded["Powers_split"].dropna().unique()
for col in columns:
    df_powers[col] = df_powers["Powers"].str.contains(col)

df_powers.head()
```

## Combining the Two Dataframes
---

```python
## Combining two dataframes
df_final = pd.merge(df_superheros, df_powers, left_on = "Hero", right_on = "hero_names")
df_final.head()
```

```python
## Testing new dataframe

## Comparing average height of those with super speed to heros without
filter_speed = df_final["Super Speed"] == True
df_speed = df_final.loc[filter_speed, :]
df_speed["Height"].mean()
```

```python
## Comparing average height of those with super speed to heros without
filter_nospeed = df_final["Super Speed"] == False
df_nospeed = df_final.loc[filter_nospeed, :]
df_nospeed["Height"].mean()
```

```python
publishers = df_final["Publisher"].unique()
publishers
```

```python
## Looping through to get average publisher height
publisher_height = []
for publisher in publishers:
    filter_pub = df_final["Publisher"] == publisher
    df = df_final.loc[filter_pub, :]
    x = df["Height"].mean()
    add = {publisher : x}
    publisher_height.append(add)

publisher_height
```
